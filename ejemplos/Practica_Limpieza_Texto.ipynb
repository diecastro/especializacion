## 1) Normaliza a minúsculas

En el DataFrame `df_texto`, crea una nueva columna `texto_min` que contenga el texto de la columna `texto` en minúsculas.

````python
# Copie su código aca
from matplotlib import table


df_texto["texto_min"] =df_texto["texto"].str.lower()
df_texto[["texto", "texto_min"]]
````

## 2) Eliminar Espacios

Genera  la columna `texto_espacios` eliminando espacios al inicio/fin

````python
# Copie su código aca
df_texto["texto_espacios"] = df_texto["texto_min"].str.strip()
df_texto[[ "texto_espacios", "texto_min"]]
````

## 3) Puntuación

Crea `texto_sin_punct` removiendo puntuación y símbolos, conservando letras y espacios.

````python
# Copie su código aca
df_texto["texto_sin_punct"] = df_texto["texto_min"].str.replace(r'[^\w\s]', '', regex=True)
df_texto[["texto_sin_punct"]]
````

## 4) Acentos

Elimina los acentos de las vocales del texto `df_texto` y grábalo en una nueva columna `texto_sin_acentos`.

````python
# Copie su código aca
import unicodedata

df_texto["texto_sin_acentos"] = df_texto["texto_sin_punct"].apply(
    lambda x: ''.join(c for c in unicodedata.normalize('NFD', str(x)) 
                      if unicodedata.category(c) != 'Mn')
)
df_texto["texto_sin_acentos"]
````

## 5) Emojis

Crear la columna `texto_sin_emoji` quitando emojis, pictogramas y símbolos gráficos comunes.

````python
# Copie su código aca
import emoji
df_texto["texto_sin_emoji"] = df_texto["texto"].apply(
    lambda x: emoji.replace_emoji(x, replace='')
)
````

## 6) Extrae emails y URLs

En `df_texto`, crea las columnas `email` y `urls` extrayendo emails y URLs del texto original.

**Patrones sugeridos:**
- Email: `[A-Za-z0-9_.+-]+@[A-Za-z0-9-]+\.[A-Za-z0-9.-]+`
- URL: `https?://\S+`

````python
# Copie su código aca
import re

df_texto["email"] = df_texto["texto"].apply(lambda x: re.findall(r'[A-Za-z0-9_.+-]+@[A-Za-z0-9-]+\.[A-Za-z0-9.-]+', x))
df_texto["urls"] = df_texto["texto"].apply(lambda x: re.findall(r'https?://\S+', x))
df_texto[["texto", "email", "urls"]]
````

## 7) Hashtags, menciones y URLs en tweets

En `df_tweets`, crea las columnas `hashtags`, `mentions` y `urls`
extrayendo hashtags, menciones y URLs del texto del tweet.

````python
# Copie su código aca

df_tweets["urls"] = df_tweets["tweet"].apply(lambda x: re.findall(r'https?://\S+', x))
df_tweets["mentions"] = df_tweets["tweet"].str.findall(r'@\w+')
df_tweets["hastags"] = df_tweets["tweet"].str.findall(r'#\w+')
df_tweets[["tweet", "hastags", "mentions","urls"]]
````

## 8) ¿Es retuit? Limpia el prefijo

Crea la columna `es_rt` (booleano) si el tweet comienza con `RT`. Luego genera la columna `tweet_sin_rt` quitando el prefijo `RT @usuario:` del inicio.

````python
# Copie su código aca
df_tweets["es_rt"] = df_tweets["tweet"].str.startswith("RT")
df_tweets["tweet_sin_rt"] = df_tweets["tweet"].str.replace(r'^RT @\w+:\s*', '', regex=True)
df_tweets[["tweet","hastags", "mentions","urls", "es_rt", "tweet_sin_rt"]]
````

## 9) `tweet_limpio`

Crea `tweet_limpio` removiendo URLs, menciones y hashtags; además, elimina la puntuación sobrante, quita espacios y pasa a minúsculas.

````python
# Copie su código aca
df_tweets["tweet_limpio"] = (
    df_tweets["tweet"]
    .str.replace(r'https?://\S+', '', regex=True)
    .str.replace(r'^RT @\w+:\s*', '', regex=True)
    .str.replace(r'@\w+', '', regex=True)
    .str.replace(r'#\w+', '', regex=True)
    .str.replace(r'[^\w\s]', ' ', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
    .str.lower()
)
df_tweets[["tweet", "tweet_limpio"]]
````

## 10) Talla en `df_productos`

Extrae y estandariza la talla:
- Letras: `XS, S, M, L, XL, XXL`, o `TU` (talla única).
- Números: captura como `talla_num`.
Crea la columna`talla_std` (letras) y `talla_num` (numérica).

````python
# Copie su código aca
import re

def extraer_talla_std(s):
    if pd.isna(s): return float('nan')
    s_low = s.lower()
    if 'única' in s_low or 'unica' in s_low: return 'TU'
    m = re.search(r'(?:talla\s+)?(XS|S|M|L|XL|XXL)\b', s, re.I)
    if m: return m.group(1).upper()
    m = re.search(r'-(XS|S|M|L|XL|XXL)\b', s, re.I)
    if m: return m.group(1).upper()
    m = re.search(r'(?:talla\s+)?([sml])(?:\s|$|\)|,)', s, re.I)
    if m: return m.group(1).upper()
    m = re.search(r'(?:^|\s)([sml])(?:\s|$)', s, re.I)
    if m: return m.group(1).upper()
    return float('nan')

def extraer_talla_num(s):
    if pd.isna(s): return float('nan')
    m = re.search(r'\d+', s)
    return float(m.group()) if m else float('nan')

df_productos["talla_std"] = df_productos["producto"].apply(extraer_talla_std)
df_productos["talla_num"] = df_productos["producto"].apply(extraer_talla_num)
df_productos[["producto", "precio", "talla_std", "talla_num"]]
````

## 11) Extraer de precios

Convierte la columna `precio` a numérico (`precio_num`) independiente del formato (`$1.234,50`, `USD 45`, `30,00 €`, `COP 9.990`, etc.).

````python
# Copie su código aca
import re

def parse_precio(s):
    s = str(s).strip()
    s = re.sub(r'[$€£]|USD|COP|AR\$|S/\.', '', s, flags=re.I)
    s = re.sub(r'\s+', '', s)
    if not s or not re.search(r'\d', s): return float('nan')
    if ',' in s and '.' in s:
        s = s.replace('.', '').replace(',', '.')
    elif ',' in s:
        s = s.replace(',', '.')
    try:
        return float(s)
    except ValueError:
        return float('nan')

df_productos["precio_num"] = df_productos["precio"].apply(parse_precio)
df_productos[["producto", "precio", "talla_std", "talla_num", "precio_num"]]
````

## 12) Filtrado por texto

Filtra el DataFrame `df_productos` para mostrar solo filas cuyo `producto` contenga **camisa** o **pantalón**, ignorando acentos y mayúsculas/minúsculas.

````python
# Copie su código aca
import unicodedata

def quitar_acentos(texto):
    nfd = unicodedata.normalize('NFD', str(texto))
    return ''.join(c for c in nfd if unicodedata.category(c) != 'Mn')

producto_norm = df_productos["producto"].apply(lambda x: quitar_acentos(x).lower())
mask = producto_norm.str.contains("camisa|pantalon", regex=True)
df_productos[mask]
````
